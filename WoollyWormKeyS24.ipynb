```python
#Import numpy
import numpy as np
TOTALDISTANCE = 42
INTERVAL = 10
TOTALTIME = 90
```

```python
#Generate a 2D array to describe the distance the woolly worm travels in 10 sec (max of 8 in/10 sec)
dist = np.random.randint(80, size = (12,9))
dist = dist / INTERVAL
dist
```

```python
#Compute the cumulative distance for each worm at each interval of 10 sec (axis = 1 means across the columns)
totaldist = dist.cumsum(axis = 1)
totaldist
```

```python
#Import matplotlib and plot the cumulative distances for each worm  by row
import matplotlib.pyplot as plt
plt.plot(totaldist.T)
```

```python
#Compute the average speed for each worm (total distance/total time)
avgspeed = totaldist[:,-1]/TOTALTIME
avgspeed
```

```python
#Compute the average speed over each 10 sec time interval (ds/dt)
speed10 = dist/INTERVAL
speed10
```

```python
#Replace any 0.0 elements with 0.01 to prevent division by 0 in next step
speed10[speed10 == 0] = 0.01
speed10
```

```python
#Which worm won? Remove all doubt by calculating the time to cross the finish line (42 in)
#The time to reach 42 inches = (42 - distance after 80 sec)/average speed in last 10 seconds
lastinterval = (TOTALDISTANCE - totaldist[:,-2])/speed10[:, -1]
lastinterval
```

```python
#Confirm max speed (expect .8 in/sec)
speed10max = np.max(speed10[:,:])
speed10max
```

```python
#Confirm min speed (expect 0.0 in/sec)
speed10min = np.min(speed10[:,:])
speed10min
```

```python
#Calculate total time of winner
wtime = 80 + np.min(lastinterval)
wtime
```
