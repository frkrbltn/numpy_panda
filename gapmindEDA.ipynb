```python
#import pandas and numpy
import numpy as np
import pandas as pd
```

```python
#Read in csv data to create DataFrame (NOTE: this is a tsv file)
#from https://github.com/chendaniely/pandas_for_everyone/blob/master/data/gapminder.tsv
df = pd.read_csv('gapminder.tsv', sep = '\t')
#Just run this cell
```

```python
#View the last row using the loc method. Make it general for any DataFrame.
df.loc[df.shape[0]-1]

# What is the global average life expectancy for the year 2002? (Round to one decimal value.)

year_2002 = df[df['year'] == 2002]
round(year_2002['lifeExp'].mean(), 1)
```

```python
#Group the entire df by year and find the average life expentancy based upon year
df.groupby('year')['lifeExp'].mean()
```

```python
#Import matplotlib and graph the average life expectancy based upon year
import matplotlib.pyplot as plt
df.groupby('year')['lifeExp'].mean().plot()
```

```python
# Come up with a good title and x and y labels for the graph 
# Make this graph look professional
# You can create histogram or bar graph
# Create the gprah based upon the pop and gdpPercap columns

df.groupby('year')['lifeExp'].mean().plot(kind = 'bar', title = 'Average Life Expectancy by Year', xlabel = 'Year', ylabel = 'Life Expectancy')
```

```python
# Create another plot based on the pop and gdpPercap columns
# following that structure: tips.plot.scatter(x = 'total_bill', y = 'tip')
df.plot.scatter(x = 'pop', y = 'gdpPercap')
```

```python
# Create another table
# Box plot â€“ illustrates the distribution with the box spanning the 1st through 3rd quartiles, line marking the
# median, outliers marked by circles.

df.boxplot(column = 'lifeExp', by = 'year')
```

```python
# Now create subplots for the above 3 graphs
# Create a 2x2 grid of subplots
# Create the first plot based upon the average life expectancy by year
plt.subplot(2, 2, 1)
df.groupby('year')['lifeExp'].mean().plot(kind = 'bar', title = 'Average Life Expectancy by Year', xlabel = 'Year', ylabel = 'Life Expectancy')

# Create the second plot based upon the pop and gdpPercap columns
plt.subplot(2, 2, 2)
df.plot.scatter(x = 'pop', y = 'gdpPercap')

# Create the third plot based upon the lifeExp column
plt.subplot(2, 2, 3)
df.boxplot(column = 'lifeExp', by = 'year')
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))  # Adjust the size as needed

# Plot 1: Average Life Expectancy by Year (Bar Chart)
lifeExp_by_year = df.groupby('year')['lifeExp'].mean()
axs[0].bar(lifeExp_by_year.index, lifeExp_by_year.values)
axs[0].set_title('Average Life Expectancy by Year')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Life Expectancy')

# Plot 2: Scatter Plot of Population vs. GDP per Capita
axs[1].scatter(df['pop'], df['gdpPercap'])
axs[1].set_title('Population vs. GDP per Capita')
axs[1].set_xlabel('Population')
axs[1].set_ylabel('GDP per Capita')

# Plot 3: Boxplot of Life Expectancy by Year
df.boxplot(column='lifeExp', by='year', ax=axs[2], grid=False)
axs[2].set_title('Life Expectancy by Year')
axs[2].set_xlabel('Year')
axs[2].set_ylabel('Life Expectancy')
axs[2].tick_params(axis='x', rotation=90)  # Rotate x labels for better visibility

plt.tight_layout()  # Adjust the layout to make sure everything fits without overlap

plt.show()
```
