Using NumPy for arrays and math operations, and Matplotlib for plotting.

Add constant for calculating electric  potential and electric field.

```python
import numpy as np
import matplotlib.pyplot as plt

# Colomb's constant
K = 9 * 10**9

# Charge of the electron
q1 = 1e-9
```

Create two identical 1D ndarrays, x and y, to hold grid values from -.5 m up through .5 m with an
interval of 0.01 m.

```python
x = np.arange(-0.5, 0.5, 0.01)
y = np.arange(-0.5, 0.5, 0.01)
```

Create a NumPy meshgrid to create a grid of points using the x and y arrays.

```python
X, Y = np.meshgrid(x, y)
```

Position particle 1 at (0.105, 0.105) with a charge of 1 nanocoulombs. Offsetting the particle from a grid
point will prevent any divide by zero errors.

```python
# Distance from particle formula
r1 = np.sqrt((X - 0.105)**2 + (Y - 0.105)**2)
V1 = K * q1 / r1

# Counter Map fpr Particles Electric Potential
plt.figure(figsize=(6, 6))
plt.contourf(X, Y, V1, levels=50, cmap='viridis')
plt.colorbar(label='Electric Potential (V)')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Electric Potential due to a 1nC Charge')
plt.plot(0.105, 0.105, 'ro')  # Marking the charge position
plt.show()
```

```python
# Repeat the above steps for particle 2, located at (0.305, -0.305) with a charge of 2 nanocoulom
q2 = 2e-9  # charge in Coulombs
r2 = np.sqrt((X - 0.305)**2 + (Y + 0.305)**2)  # distance from particle 2
V2 = K * q2 / r2

# Total electric potential
V_total = V1 + V2

plt.figure(figsize=(6, 6))
plt.contourf(X, Y, V_total, levels=50, cmap='viridis')
plt.colorbar(label='Electric Potential (V)')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Total Electric Potential due to Two Charges')
plt.plot([0.105, 0.305], [0.105, -0.305], 'ro')  # Marking the charge positions
plt.show()
```
